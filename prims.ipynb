```python
# generate adjecency matrix
import random

def generateRandomGraph(dimension):
    result = list()
    for i in range(dimension):
        result.append([0] * dimension)
    for i in range(dimension):
        for j in range(i+1,dimension):
            if random.randint(0,1):
                result[i][j] = result[j][i] =random.randint(0,9)
                    
                
                
    return result

adjecencyMatrix = generateRandomGraph(10)
for i in adjecencyMatrix:
    print(i)

```

```python
import networkx as nx
import matplotlib.pyplot as plt

def visualizeGraph(adjMatrix):
    graph = list()
    for i in range(len(adjMatrix)):
        for j in range(i+1,len(adjMatrix)):
            if adjMatrix[i][j] :
                graph.append([i,j])
    G = nx.Graph()
    G.add_edges_from(graph)
    nx.draw_networkx(G)
    plt.show()
    print(graph)

visualizeGraph(adjecencyMatrix)
```
